```python
#grid
LX =320
LT =320
LI =12
dx =1
```

```python
#parameter
#a
Da =0.008
#ra =0.04 #|||
#ra =0.08 #---
ra = np.linspace(0.03,0.08,LX+1)
ba =0.05
sa =0.0
#b
Db =0.35
rb =0
bb =0.03  #####
s  =ra

# Db =0.00035
# #initial value
# bb     =random.random(LX+1)
# for i in range(int(LX+1)):
#     if(bb[i]>0.9):
#         bb[i]=0.05
#     else:
#         bb[i]=0.04
```

```python
#initial value
a     =random.random(LX+1)+2.7
b     =random.random(LX+1)+0.7
Ma    =np.zeros((LT,LX+1))
Mb    =np.zeros((LT,LX+1))
```

```python
#diffuse
def d2_dx2(Z): 
    Zleft = np.roll(Z, 1, axis=0)
    Zright = np.roll(Z, -1, axis=0)
    return (Zleft + Zright   - 2.0 * Z) / dx**2
```

```python
#simulation
for j in range(int(LT)):
    for i in range(int(LI)):
        a, b = \
        a + ( s*(a*a/(1+sa*a*a)+ba)*b - ra*a + Da*d2_dx2(a) ),\
        b + ( bb-s*b*(a*a/(1+sa*a*a)+ba) - rb*b + Db*d2_dx2(b) )
    Ma[j,:] = a
    Mb[j,:] = b
```

```python
from matplotlib import ticker
fig, ax = plt.subplots(1, 2, figsize=(16, 6))
im0 = ax[0].imshow(Ma,cmap='YlGn');#, clim=(0,0.1));
ax[0].set_title('A density g/m$^2$');
cbar=plt.colorbar(im0, ax=ax[0],fraction=0.046,pad=0.04);
tick_locator = ticker.MaxNLocator(nbins=5)
cbar.locator = tick_locator
cbar.update_ticks()

im1 = ax[1].imshow(Mb,cmap='YlGn');#, clim=(0,0.5));
ax[1].set_title('B density g/m$^2$');
cbar=plt.colorbar(im1, ax=ax[1],fraction=0.046,pad=0.04);
tick_locator = ticker.MaxNLocator(nbins=5)
cbar.locator = tick_locator
cbar.update_ticks()
plt.savefig('4.par-ra-1.png', dpi=350)
```
